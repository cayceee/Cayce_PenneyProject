```python
### DATA GENERATION WITH STATE ###

# FUNCTIONS #

import random
from datetime import datetime as dt
from typing import Callable
import os


DEBUG = True
def debugger_factory(show_args=True) -> Callable:
    def debugger(func:Callable) -> Callable:
        if DEBUG:
            def wrapper(*args, **kwargs):
                if show_args:
                    print(f'{func.__name__} was called with:')
                    print('Positional arguments:\n', args)
                    print('Keyword arguments:\n', kwargs)

                t0 = dt.now()
                results = func(*args, **kwargs)
                print(f'{func.__name__} ran for {dt.now()-t0}')
                return results
        else:
            return func
        return wrapper
    return debugger



def initial_deck(n = 26):
  """
  Creates the initial, unshuffled deck, with 0's and 1's representing black and red cards.
  The values of the cards are arbitrary with respect to the game.
  The deck is returned as 'initial_deck', a list.
  """

  initial_deck = [0]*n + [1]*n
  return initial_deck



@debugger_factory()



def makeDecks_withState(newState, v = 100000):
    """
    Makes a v amount of decks, set to 100,000 by default.
    Presents 'decks' as a tuple of the deck lists.
    Uses a presestablish system state as variable 'newState'.
    """

    random.setstate(newState)
    decks = []
    for i in range(v):
      newDeck = initial_deck()
      random.shuffle(newDeck)
      decks.append(newDeck)

    return decks



# Sample Main for Code Generation

newState = (3, (2147483648, 2256425436, 1801895169, 940611285, 1016410359, 2709328919, 450659928, 1006268007, 1300019232, 375853835, 3523341169, 3724058608, 3996584191, 1891394466, 394835466, 3101348516, 3303513004, 1796572336, 2135702550, 4202310974, 1355404475, 1485098653, 2609662263, 110502034, 3517778678, 2963013606, 2900485339, 2403372541, 1062936140, 1675171944, 1088257967, 2205003211, 1777544203, 3011588845, 1600546379, 2231295115, 3099536231, 2499558228, 506270477, 2753472204, 2247202614, 2745852132, 3888511959, 3995555072, 1922631207, 3416218076, 214100577, 555208435, 2939295898, 2209409741, 3934801211, 695108287, 656557445, 3435044399, 2583975767, 845511151, 2619266600, 42840239, 1725879619, 1215985260, 1361874730, 2310575393, 2984364263, 1654028797, 1960063467, 1924296495, 446347475, 1138088743, 3101552666, 3494382100, 721601582, 3402274196, 2631097065, 998066076, 643427757, 1127045738, 2996308770, 2101880747, 3248940247, 4247950846, 3099014464, 3394344599, 1310215483, 1835481206, 4176566295, 370968811, 1257147581, 2628304739, 2685005105, 100547934, 1626743719, 3316319996, 2790456147, 2854812674, 2733824104, 360312068, 1300675336, 1019255004, 2117868532, 3435188150, 3478516183, 1058396954, 1977749074, 680397786, 1789440413, 3820148218, 642028311, 1339512257, 2422684407, 3769469079, 1815842728, 3454504129, 3679167515, 4238021687, 2836200538, 2604554883, 4057718172, 1407455111, 2250854588, 4227442302, 3678670787, 1841667519, 3858240130, 3590885453, 1878903278, 1777742535, 4203194738, 2728420071, 471370730, 3181082062, 2034506399, 1949132176, 1885513894, 3537036234, 863030982, 603436644, 1450845645, 3305842569, 3346193590, 1522424229, 66523425, 2011881450, 1755276609, 1950545862, 2849949036, 2234575768, 38337010, 3306532703, 2767261659, 2262209860, 2935940945, 3539748351, 2982462636, 3914873141, 3724040457, 3096826360, 2558848610, 2319985031, 2253127358, 822779461, 3173553940, 1591943662, 1927954463, 4220134594, 2074238999, 3189242494, 584321780, 160993454, 4117417565, 386956109, 4259055415, 1021436380, 467196664, 824677608, 947029794, 2413767207, 751279018, 131984369, 3108812007, 1391486070, 2856881385, 344048793, 1433299700, 2714862769, 93038881, 2993323684, 3952263404, 2881302148, 3642965941, 2883785090, 2675554772, 2328806970, 2257385306, 1813416541, 4008901212, 703163008, 3582719452, 3105182468, 484674475, 28300111, 3531348000, 3853933230, 2673437793, 2401146601, 183779102, 2348771633, 2239286854, 2700243382, 624331533, 1589294725, 3893064913, 2539392995, 1404563788, 552183342, 704861262, 725367240, 1573568665, 582560201, 490250574, 2810765803, 2274431379, 3955122909, 2643844892, 3764575806, 628864781, 1719174902, 1012035069, 2537489562, 1110549365, 1789207469, 1029396746, 2281500030, 2902674902, 153364761, 3315309083, 2350832515, 1077503787, 2614198812, 4271976685, 2112543459, 2664507071, 1537974500, 611277822, 3717980507, 2648781091, 3462311640, 1179288063, 4267597891, 2154683995, 1335891197, 2002058509, 2556375011, 3671006822, 2528128075, 1176777830, 2994610046, 2902909480, 4266966422, 566227726, 3295138542, 770392502, 833471862, 3641979400, 644268371, 2980914509, 860935751, 3279279308, 198084803, 2430317831, 1562285984, 1431588559, 3583139024, 3781666547, 3468258784, 653651595, 3867026508, 575047248, 3672894103, 2044110471, 2657706683, 3042328585, 1663831556, 3948635162, 2160685823, 4181254187, 3041240509, 1500267863, 2900471124, 3085856846, 2370579206, 527069948, 558515105, 4287113154, 3289436556, 2609106357, 3090688294, 3978822208, 21231399, 648698632, 2123651044, 381823350, 1165173230, 3897584758, 3006554094, 3262807176, 2440048321, 4235825514, 1793569979, 28470923, 355219719, 223028386, 4109276459, 3836887115, 4218347370, 579751986, 1479719121, 1193093292, 803099979, 2074123494, 3241797902, 3695648148, 3477825570, 4132456077, 834893975, 2685914798, 1490639753, 407570183, 1097525524, 2555875815, 3750175086, 3924025475, 1113127973, 3731354971, 3089750115, 3593777374, 3642720078, 1464487380, 2696023059, 2606544724, 2631228536, 3105625728, 3542073518, 1283962489, 3971449029, 3727267028, 393072684, 2915160084, 3227530274, 2208391959, 1753881306, 1322895958, 2342374676, 960657276, 2766129033, 913825623, 1226684758, 4121944791, 1347960978, 1704910794, 3586852075, 1709008115, 853101538, 2712899931, 2962943256, 4113778493, 657479878, 774549052, 1634184713, 157877642, 1347984858, 2666643809, 1082220565, 1815308356, 2056861348, 1719238459, 2636659550, 355357900, 2244639954, 1547287143, 32359081, 3080017553, 2083225855, 2071985933, 369761338, 1770753240, 3068339319, 632915815, 3970472545, 1019886634, 982212903, 3656632798, 4099977280, 1733680096, 3510409855, 534969654, 1001567036, 2813744925, 1392095989, 131520902, 1380416526, 2107704047, 1449061743, 208976488, 186854464, 1960237855, 547085742, 4159243272, 3732928283, 474863341, 1049852367, 1984141511, 3174711004, 3961310166, 479921557, 2657083008, 925727572, 2781412338, 1400372851, 2210455993, 2237770884, 1696689255, 293101053, 865694237, 1754863966, 4170278896, 3499037306, 401785217, 2140288615, 2057667468, 3841147723, 3759949089, 3011867921, 1842624406, 476140999, 883602268, 1877468256, 1123478719, 4047857447, 1037566171, 4196987155, 1306654679, 1893179224, 941367569, 489379859, 3473058654, 3005430740, 1659269452, 226425692, 2644686816, 2282214236, 2164504302, 39064465, 4152309555, 4210253747, 2998879468, 2044029159, 1750417150, 379149034, 551848060, 2863565809, 3371726009, 3260027684, 3010135846, 2972648287, 2804904753, 3061276854, 2034936464, 2952131038, 1572547571, 3918311441, 1192019349, 1376985667, 1067151005, 1369967751, 722334660, 3763058309, 2789259151, 2235338635, 4274896593, 2252528840, 118080173, 1696842075, 147373729, 1758985398, 4253856497, 671904191, 1212878776, 3122971235, 1570996825, 85275751, 496960253, 182713102, 1779929407, 2520400998, 1766439022, 3362838310, 3609123655, 4127436924, 793236575, 273365495, 2418045551, 2191663120, 1656432334, 1881735228, 3967637361, 1129925840, 204239488, 3775949549, 3854554609, 3256870970, 624770443, 2014468814, 321309600, 957349113, 3356455402, 2956905052, 2464832197, 3710603098, 3131716091, 1740312227, 2790448125, 2387903432, 3594825490, 2590027395, 2044526153, 3626046615, 589924068, 3869712302, 295827964, 1525357498, 3439427546, 4115038694, 3044465926, 17283411, 2366549319, 1524004432, 3787772950, 410638793, 3227478143, 1608322056, 3622761921, 1563502060, 2093168893, 9354604, 1622806275, 3379022753, 309903922, 2456827314, 2528477449, 467843358, 2893471902, 3941895455, 1473296123, 1316115943, 3396488586, 1283101970, 2324722682, 3725667661, 385155394, 150191163, 2983433868, 2085203638, 895319151, 3463214500, 675738265, 2646487448, 4133546174, 1904829944, 3233016340, 3032863674, 2243445950, 547172515, 944849194, 1774697171, 488895727, 2913257645, 2483115875, 2097291495, 1631331890, 4113252914, 1641107393, 2271147388, 2943940936, 3131666985, 2759243197, 3959840523, 2799059919, 2825974578, 419722777, 2020975803, 2769219867, 1058173810, 2451525736, 1927487092, 1688118032, 2630143334, 382706481, 3956209430, 2821307749, 3918426349, 1157703851, 59866782, 1736396112, 1911053328, 1633129536, 41680204, 3215396943, 511729163, 3924669169, 706744663, 2534948114, 2216802413, 648806714, 660343093, 3923292247, 1874699129, 4137459738, 4068160901, 2270527869, 1938099927, 3034790456, 2843607762, 1931023239, 1495297380, 624), None)
# Replace newState with a copy a pasted tuple representing the system state you want to run the simulation with.
# Mostly used for reproducability

decks = makeDecks_withState(newState)
print("First deck:", decks[0])
print("Last deck:", decks[99999])
```

```python
### RUNNING SIMULATION ###

def playerNotKnowChoiceFUNCTION():
  """
  Produces a list of three random numbers between 0 and 1.
  Represents the player who is not in the know of the game's trick.
  """
  playerNotKnowChoice = []
  playerNotKnowChoice.append(random.randint(0,1))
  playerNotKnowChoice.append(random.randint(0,1))
  playerNotKnowChoice.append(random.randint(0,1))

  return playerNotKnowChoice



def playerKNOWChoiceFUNCTION(playerNotKnowChoice):
  """
  Produces a list of numbers based off of the playerNotKnowChoice.
  Represents the player who is in the know of the game's trick.
  The formula for the playerKNOWChoice is [Opposite of player Not Know's second number, player Not Know's first number, player Not Know's second number]
  """
  playerKNOWChoice = []
  if playerNotKnowChoice[1] == 0:
    playerKNOWChoice.append(1)
  else:
    playerKNOWChoice.append(0)

  playerKNOWChoice.append(playerNotKnowChoice[0])
  playerKNOWChoice.append(playerNotKnowChoice[1])

  return playerKNOWChoice



def dealOne(deck):
  """
  Given a deck, this function removes the first card from the deck and returns it.
  """
  card = deck[0]
  deck.remove(deck[0])

  return card



def play(playerNotKnowChoice, playerKNOWChoice, deck):
  """
  Plays through one deck of Penney's Game
  Score is kept from number of stacks of cards won, not number of cards won.
  """
  dealtCards = []
  playerNotKnowScore = 0
  playerKNOWScore = 0
  ties = 0

  while len(deck) > 0:
    dealtCards.append(dealOne(deck))
    if len(dealtCards) >= 3:
      checkCards = dealtCards[len(dealtCards) -3 : len(dealtCards)]
      if checkCards == playerNotKnowChoice:
        playerNotKnowScore = playerNotKnowScore + 1
        dealtCards = []
      elif checkCards == playerKNOWChoice:
        playerKNOWScore = playerKNOWScore + 1
        dealtCards = []
      elif len(deck) == 0:
        ties = ties + 1

  return playerNotKnowScore, playerKNOWScore, ties



def simulate(decks):
  """
  Plays the game, resetting with each deck.
  Wins refer to the amount of decks that player won.
  """

  playerNotKnow_WINS = 0
  playerKNOW_WINS = 0
  gameTies = 0

  for deck in decks:
    current_playerNotKnowChoice = playerNotKnowChoiceFUNCTION()
    current_playerKNOWChoice = playerKNOWChoiceFUNCTION(current_playerNotKnowChoice)
    playerNotKnowScore, playerKNOWScore, ties = play(current_playerNotKnowChoice, current_playerKNOWChoice, deck)
    if playerNotKnowScore > playerKNOWScore:
      playerNotKnow_WINS = playerNotKnow_WINS + 1
    elif playerNotKnowScore < playerKNOWScore:
      playerKNOW_WINS = playerKNOW_WINS + 1
    elif playerNotKnowScore == playerKNOWScore:
      gameTies = gameTies + 1

  return playerNotKnow_WINS, playerKNOW_WINS, gameTies



### Example of Running the Simulation ###

simulationResults = simulate(decks)
print("________________________________________________________")
print()
print(simulationResults)
print("________________________________________________________")
print("Player Who Did Not Know the Trick Won", simulationResults[0], "times")
print("Player Who Did Know the Trick Won", simulationResults[1], "times")
print("There were", simulationResults[2], "ties")
```
